# Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
```

# Step 2: Create Dataset

```python
X = np.array([[1],
              [2],
              [3],
              [4],
              [5]])

y = np.array([3,5,7,9,11])
```

# Step 3: Train with Learning Rate = 0.01srt

```python
model = SGDRegressor(
    learning_rate='constant',
    eta0=0.01,
    max_iter=1000,
    random_state=42
)

model.fit(X, y)

prediction = model.predict(X)

print("Weight :", model.coef_[0])
print("Bias :", model.intercept_[0])

mse = mean_squared_error(y, prediction)

print("MSE :", mse)
```

# Step 4: Test Multiple Learning Rates

```python
learning_rates = [0.0001, 0.001, 0.01, 0.1]

for lr in learning_rates:

    model = SGDRegressor(
        learning_rate='constant',
        eta0=lr,
        max_iter=1000,
        random_state=42
    )

    model.fit(X, y)

    prediction = model.predict(X)

    mse = mean_squared_error(y, prediction)

    print("--------------------------------")
    print("Learning Rate :", lr)
    print("Weight :", model.coef_[0])
    print("Bias :", model.intercept_[0])
    print("MSE :", mse)
```

# Step 5: Plot Predictions

```python
plt.scatter(X, y, color="blue", label="Actual")

plt.plot(X, prediction, color="red", label="Prediction")

plt.xlabel("Experience")
plt.ylabel("Salary")

plt.legend()

plt.show() 
```

# Step6: Compare Learning Rates in One Program

```python
learning_rates = [0.0001, 0.001, 0.01, 0.1]

for lr in learning_rates:

    model = SGDRegressor(
        learning_rate="constant",
        eta0=lr,
        max_iter=1000,
        random_state=42
    )

    model.fit(X, y)

    prediction = model.predict(X)

    mse = mean_squared_error(y, prediction)

    print(f"Learning Rate : {lr}")
    print(f"Weight        : {model.coef_[0]:.4f}")
    print(f"Bias          : {model.intercept_[0]:.4f}")
    print(f"MSE           : {mse:.6f}")
    print()
```

# Step 7: Experiment with Epochs

```python
epochs = [10,100,500,1000]

for e in epochs:

    model = SGDRegressor(
        learning_rate="constant",
        eta0=0.01,
        max_iter=e,
        random_state=42
    )

    model.fit(X,y)

    prediction = model.predict(X)

    mse = mean_squared_error(y,prediction)

    print("Epochs :", e)
    print("Weight :", model.coef_[0])
    print("Bias :", model.intercept_[0])
    print("MSE :", mse)
    print()
```

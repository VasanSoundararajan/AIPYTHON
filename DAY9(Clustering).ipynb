# Cell 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

# Cell 2: Create Dataset

```python
data = {
    "Age":[22,24,25,27,45,46,48,50,60,62],
    "Income":[25000,27000,30000,32000,70000,72000,75000,78000,100000,105000]
}

df = pd.DataFrame(data)

df
```

# Cell 3: Feature Scaling

```python
scaler = StandardScaler()

X = scaler.fit_transform(df)
```

# Cell 4: Train K-Means

```python
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

kmeans.fit(X)
```

# Cell 5: Assign Cluster Labels

```python
df["Cluster"] = kmeans.labels_

df
```

# Cell 6: Cluster Centers

```python
print(kmeans.cluster_centers_)
```

# Cell 7: Inertia (WCSS)

```python
print("Inertia:", kmeans.inertia_)
```

# Cell 8: Silhouette Score

```python
score = silhouette_score(X, kmeans.labels_)

print("Silhouette Score:", score)
```

# Cell 9: Predict a New Customer

```python
new_customer = [[35,50000]]

new_customer = scaler.transform(new_customer)

cluster = kmeans.predict(new_customer)

print("Cluster:", cluster[0])
```

# Cell 10: Visualize Clusters

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df["Age"],
    df["Income"],
    c=df["Cluster"],
    s=100
)

plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Customer Clusters")

plt.show()
```

# Cell 11: Elbow Method

```python
wcss = []

for k in range(1,11):

    model = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    model.fit(X)

    wcss.append(model.inertia_)

plt.plot(range(1,11), wcss, marker="o")

plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS")

plt.title("Elbow Method")

plt.show()
```

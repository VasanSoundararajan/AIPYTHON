```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

# Step 2: Create Sample Dataset

```python
data = {
    "Name":["John","Alice","Bob","David","Emma"],
    "Department":["IT","HR","Finance","IT","HR"],
    "Gender":["Male","Female","Male","Male","Female"],
    "Age":[25,30,28,35,27],
    "Salary":[50000,60000,55000,75000,58000]
}

df = pd.DataFrame(data)

df
```

# Label Encoding

Use Label Encoding when categories have an order or when you're encoding the target variable. Although you can apply it to nominal data, it may introduce an artificial ordering.

```python
le = LabelEncoder()

df["Department_Label"] = le.fit_transform(df["Department"])

df
```

```python
print(le.classes_)
```

```python
gender_encoder = LabelEncoder()

df["Gender_Label"] = gender_encoder.fit_transform(df["Gender"])

df
```

# One-Hot Encoding

For nominal categories like Department, One-Hot Encoding is usually preferred.

```python
department_encoded = pd.get_dummies(
    df["Department"],
    prefix="Dept"
)

department_encoded
```

```python
df = pd.concat([df, department_encoded], axis=1)

df
```

# Remove Original Column

After encoding, the original categorical column is often dropped.

```python
df = df.drop("Department", axis=1)

df
```

# Standardization
| Age | Salary |
| --- | ------ |
| 25  | 50000  |
| 30  | 60000  |

The values are on different scales.

We standardize them.

```python
scaler = StandardScaler()
```

```python
df[["Age","Salary"]] = scaler.fit_transform(
    df[["Age","Salary"]]
)

df
```

```python
print(df["Age"].mean())
```

```python
print(df["Age"].std())
```

# Standardize Multiple Columns

```python
numerical_columns = ["Age","Salary"]

scaler = StandardScaler()

df[numerical_columns] = scaler.fit_transform(
    df[numerical_columns]
)
```

```python
df
```

```python
print(df["Age"].mean())
print(df.Salary.mean())
```

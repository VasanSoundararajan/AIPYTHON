# Objective

Predict Salary from Experience using:

* Linear Regression
* Mean Squared Error (MSE)
* Gradient Descent
* Different Learning Rates

# Step 1: Import Library

```python
import numpy as np
```

# Step 2: Create Dataset

```python
X = np.array([1,2,3,4,5])
Y = np.array([3,5,7,9,11])
```

# Step 3: Initialize Parameters

```python
weight = 0
bias = 0
learning_rate = 0.01 # good learning_rate 0.001 slow 0.0001 very slow 1 fast 10 very fast 
epochs =1000
n = len(X)
```

# Step 4: Gradient Descent Algorithm

```python
for epoch in range(epochs):
    Y_pred = weight * X + bias
    dw = (-2/n) * np.sum(X * (Y - Y_pred))
    db = (-2/n) * np.sum(Y - Y_pred)
    weight = weight - learning_rate * dw
    bias = bias - learning_rate * db
    if epoch % 500 == 0:
        mse = np.mean((Y - Y_pred)**2)
        print(epoch, mse, weight, bias)
```

# Step 5: Display Final Values

```python
print("Weight :", weight)

print("Bias :", bias)
```

# Step 6: Test Prediction

```python
experience = 6

salary = weight * experience + bias

print(salary)
```

# Step 7: Calculate MSE

```python
Y_pred = weight * X + bias

mse = np.mean((Y - Y_pred)**2)

print(mse)
```

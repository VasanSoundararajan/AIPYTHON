```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Genre'] = encoder.fit_transform(df['Genre'])
df['Platform'] = encoder.fit_transform(df['Platform'])
df['Publisher'] = encoder.fit_transform(df['Publisher'])
```

# Visualization

```python
top10 = df.head(n=10)
width = 0.1
x = np.arange(len(top10))
plt.figure(figsize=(12,5))

plt.bar(x - width, top10["NA_Sales"], width, label="NA")
plt.bar(x, top10["EU_Sales"], width, label="EU")
plt.bar(x + width, top10["JP_Sales"], width, label="JP")

plt.xticks(x, top10["Name"], rotation=90)

plt.legend()

plt.show()
```
